```python
# -*- coding: utf-8 -*-
"""
Created on Tue Apr 26 11:05:32 2022

@author: Gacemi Mohamed el Amine
Centre des Techniques Spatiales
"""

#import required libraries
import matplotlib.pyplot as plt
import geopandas as gpd
import rasterio
from rasterio.plot import show
print("ok")
pointData = gpd.read_file('/content/drive/MyDrive/Classroom/Python Langage de Programmation/points.shp')
print(pointData.crs)
pointData.plot()

ndviRaster = rasterio.open('/content/drive/MyDrive/Classroom/Python Langage de Programmation/mnt.tif')
print(ndviRaster.crs)
print(ndviRaster.count)

fig, ax = plt.subplots(figsize=(12,12))
pointData.plot(ax=ax, color='orangered')
show(ndviRaster, ax=ax)

for point in pointData['geometry']:
    print(point.xy[0][0],point.xy[1][0])
    
for point in pointData['geometry']:
    x = point.xy[0][0]
    y = point.xy[1][0]
    row, col = ndviRaster.index(x,y)
    print("Point correspond to row, col: %d, %d"%(row,col))
    print("Raster value on point %.2f \n"%ndviRaster.read(1)[row,col])
```
